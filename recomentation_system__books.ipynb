```python
def recommend_books(book_title, model, book_bivot, books_name, final_rating, n_neighbors=7):

    try:
        book_index = np.where(book_bivot.index == book_title)[0][0]
        distances, suggestions = model.kneighbors(book_bivot.iloc[book_index,:].values.reshape(1,-1), n_neighbors=n_neighbors)

        recommended_books_info = []
        for i in range(1, len(suggestions[0])):  # Start from 1 to exclude the input book
            recommended_book_title = book_bivot.index[suggestions[0][i]]
            ids = np.where(final_rating['title'] == recommended_book_title)[0][0]
            image_url = final_rating.iloc[ids]['image_url']
            recommended_books_info.append({'title': recommended_book_title, 'image_url': image_url})

        return recommended_books_info

    except IndexError:
        return "Book not found in the dataset."

# Define the book title before calling the function
input_book_title = 'You Belong To Me'

# Call the function with the defined variable
recommendations = recommend_books(input_book_title, model, book_bivot, books_name, final_rating)

# Use the defined variable in the print statement
print(f"Recommended Books for '{input_book_title}':")
for book_info in recommendations:
    print(f"Title: {book_info['title']}")
    print(f"Image URL: {book_info['image_url']}")

```
